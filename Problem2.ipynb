```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn import metrics
from sklearn.metrics import accuracy_score

%matplotlib inline
sns.set_style('whitegrid')
```

```python
data = pd.read_csv('problem_2/train.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe(include='all')
```

```python
remove = ['srno','hubpincode','custpincode','memcreateddate','mobilepurchasedate']
data.drop(columns=remove,inplace=True)
```

```python
len(data.pickupStartdate.unique())
```

```python
data.tat_in_days.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
features = ['mobmake','hubid','pickuppartnerid','deliverypartnerid','insuranceCompanyId','custpincodecategory','claimtype','claimintdate','servicecntrid','pickupStartdate']

enc = LabelEncoder()
for col in features:
    data[col] = enc.fit_transform(data[col])
```

```python
data.head()
```

```python
X = data[features]
y = data['tat_in_days']
```

```python
X.head()
```

```python
# Splitting given training data further into train and test, in order to make accurate predictions ( 80% - 20% )
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.2)
```

```python
# Model 1
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

linreg.fit(X_train,y_train)
y_pred = linreg.predict(X_test)
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Model 2
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=31)

knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Model 3
from sklearn.ensemble import GradientBoostingRegressor
grad = GradientBoostingRegressor()

grad.fit(X_train,y_train)
y_pred = grad.predict(X_test)
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Model 4
from sklearn.ensemble import RandomForestRegressor
rand_for = RandomForestRegressor(n_estimators=50)

rand_for.fit(X_train,y_train)
y_pred = rand_for.predict(X_test)
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Model 5
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X_train)

poly_reg_model = LinearRegression()
poly_reg_model.fit(X_poly,y_train)

y_pred = poly_reg_model.predict(poly_reg.fit_transform(X_test))
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Model 6
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=0)

tree_reg.fit(X_train,y_train)
y_pred = tree_reg.predict(X_test)
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Model 7
from sklearn.linear_model import SGDRegressor

# scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_sc = scaler.transform(X_train)
X_test_sc = scaler.transform(X_test)

sgd = SGDRegressor(loss="squared_loss",average=True)
sgd.fit(X_train_sc, y_train)
y_pred = sgd.predict(X_test_sc)
err_rms = (sum((y_test-y_pred)**2)/len(y_test))**(0.5)
print(err_rms)
```

```python
# Comparing all the root mean square errors, Gradient Boosting Regressor is chosen as best fit model.
```
